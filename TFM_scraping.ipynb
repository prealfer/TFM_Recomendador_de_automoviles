## Script de scrapeo con BeautifulSoup

```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#t1=time()
main_page = rq.get("http://www.autodescuento.com")
main_soup = BeautifulSoup(main_page.content, 'html.parser')

lista_marcas = []
for marca in main_soup.find_all('h2', class_='h2'):
    
    lista_marcas.append(marca.get_text().strip().replace(' ','-'))

lista_marcas_limpia = lista_marcas[1:-2]
lista_marcas_limpia

#page = rq.get("http://www.autodescuento.com/coches-nuevos-Subaru.htm", headers={'User-Agent': 'Safari'})
#soup = BeautifulSoup(page.content, 'html.parser')
path_marcas = "http://www.autodescuento.com/coches-nuevos-"
modelos_marca = []
versiones_modelo = []
count = 0

for marca in lista_marcas_limpia:
    
    page_1 = rq.get(path_marcas+marca+".htm", headers={'User-Agent': 'Safari'})
    soup_1 = BeautifulSoup(page_1.content, 'html.parser')
    
    for modelo in soup_1.find_all('p', class_='mm'):

        #modelo_con_espacio = modelo.get_text()
        #modelo_sin_espacio = modelo_con_espacio.replace(' ','')
        modelos_marca.append(modelo.get_text().replace(' ',''))
        page_2 = rq.get(path_marcas+modelos_marca[count]+".htm", headers={'User-Agent': 'Safari'})
        soup_2 = BeautifulSoup(page_2.content, 'html.parser')
        
        for version in soup_2.find_all('div', class_='version'):

            #modelo_con_espacio = modelo.get_text()
            #modelo_sin_espacio = modelo_con_espacio.replace(' ','')
            #hay que meter concatenacion de marcas - modelo - version para el futuro Dataframe
            versiones_modelo.append(version.get_text())
        count+=1

versiones_modelo
#t2=time()
#T_total=((t2-t1)/60)
#T_total
```

```python
listado_modelos = versiones_modelo[3:]
```

```python
listado_modelos_codificado = []

for model in listado_modelos:
    listado_modelos_codificado.append(model.encode('utf-8'))
    
```
